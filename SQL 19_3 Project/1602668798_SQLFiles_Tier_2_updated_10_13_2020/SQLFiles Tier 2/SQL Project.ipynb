```python
# SQL Queries


#QUESTIONS:
#/* Q10: Produce a list of facilities with a total revenue less than 1000.
# The output of facility name and total revenue, sorted by revenue. Remember
# that there's a different cost for guests and members! */

Q10 = """ SELECT
    Sub.Facility AS "Facility Name",
    (SUM(Sub.Gross) - Sub.MaintenanceCost) AS Revenue
FROM
    (
        SELECT
            F.name AS Facility,
            F.monthlymaintenance AS MaintenanceCost,
            CASE
                WHEN B.memid = 0 THEN B.slots * F.guestcost
                ELSE B.slots * F.membercost
            END AS Gross
        FROM
            Bookings AS B
        LEFT JOIN
            Facilities AS F ON F.facid = B.facid
    ) AS Sub
GROUP BY
    Sub.Facility, Sub.MaintenanceCost
HAVING
    Revenue > 1000
ORDER BY
    Revenue DESC; """

#/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */
Q11 = """ SELECT CONCAT(M.surname, ",", M.firstname) AS Member,
    CASE
    WHEN M.recommendedby = "" THEN "N.A."
    WHEN M.recommendedby IS NULL THEN "N.A."
    ELSE CONCAT(R.surname, ",", R.firstname)
    END AS Recommenders
FROM  Members AS M
LEFT JOIN Members AS R ON M.recommendedby = R.memid
ORDER BY Member; """

#
#/* Q12: Find the facilities with their usage by member, but not guests */
Q12 =  """SELECT COUNT(B.bookid) AS MemberUsage,
        F.name AS Facility
FROM Bookings AS B
LEFT JOIN Facilities AS F ON B.facid = F.facid
WHERE B.memid != 0
GROUP BY F.name
ORDER BY MemberUsage DESC; """

#/* Q13: Find the facilities usage by month, but not guests */

Q13 = """ SELECT COUNT(B.bookid) AS MemberUsage,
        F.name AS Facility,
        STRFTIME('%m', B.starttime) AS Month
FROM Bookings AS B
LEFT JOIN Facilities AS F ON B.facid = F.facid
WHERE B.memid != 0
GROUP BY F.name, STRFTIME('%m', B.starttime) 
ORDER BY `Month` ASC """
```

```python
database = "sqlite_db_pythonsqlite.db"
def execute_and_print_query(sql_query):
    
    conn = create_connection(database) # Re-use connection function
    
    if conn is not None:
        try:
            
            conn.row_factory = sqlite3.Row 
            cur = conn.cursor()
            cur.execute(sql_query)
            rows = cur.fetchall()
            
            if rows:
                # Print column headers (from the cursor's description)
                headers = [description[0] for description in cur.description]
                print("{:<30} {:<15}".format(headers[0], headers[1] if len(headers) > 1 else ""))
                print("-" * 45)
                
                # Print the data
                for row in rows:
                    # Accessing columns by index for clean formatting
                    col1 = row[0]
                    col2 = row[1] if len(row) > 1 else ""
                    
                    # Formatting the output nicely
                    if isinstance(col2, (int, float)):
                        print("{:<30} {:<15.2f}".format(col1, col2))
                    else:
                        print("{:<30} {:<15}".format(col1, col2))

            else:
                print("No results found.")

        except Error as e:
            print(f"SQL Error during execution: {e}")
        
        finally:
            if conn:
                conn.close()
    else:
        print("Failed to establish database connection.")

```
