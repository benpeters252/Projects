```python
""" This notebook is used to upload, combine, clean, and prepare data for a MN housing cost prediction. 
Data is from United States - Redfin Housing Market (CSV from Kaggle). Data was downloaded on 11/19/2025 """
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# uploading the data sets MN-1 and MN-2

MN1 = pd.read_csv('MN-1.csv')
MN2 = pd.read_csv('MN-2.csv')
```

```python
# MN1 table sample
MN1.head()
```

```python
MN1.info()
```

```python
# MN2 table sample
MN2.head()
```

```python
MN2.info()
```

```python
# a total of 27 columns for each table. To combine, its necessary to make sure all columns are the same
total_matching_columns = sum(MN1.columns == MN2.columns)
matching_columns = (MN1.columns == MN2.columns)
print("The matching columns in order are:")
print(matching_columns)
print("The total matching columns are:", total_matching_columns)
```

```python
# all columns match and the order matches.
MN_housing = pd.concat([MN1, MN2], axis=0)
MN_housing.head()
```

```python
MN_housing.info()
```

```python
# concat was successfull, next step is to clean data
# will first look at columns with excessive amount of null values
# All houses in this df are for sale, so there is no sold date --> all are NaN
# 

MN_cleaned = MN_housing.drop(columns=['SOLD DATE'], inplace=False)
MN_cleaned.head()
```

```python
# Counting NaNs

MN_cleaned.isna().sum()
```

```python
# Removing all missing rows for Price. I expect this will remove the NaN values for other columns as well

MN_cleaned.dropna(subset=['PRICE'], inplace = True)
MN_cleaned.isna().sum()
```

```python
# All NaN values were removed from most columns. 
# the open house columns are missing 557/700 total rows (80%). Also, I do not see why this would affect housing price. Will remove columns
#  The URL will also be removed

MN_cleaned.drop(columns=['NEXT OPEN HOUSE START TIME', 'NEXT OPEN HOUSE END TIME', 'URL (SEE https://www.redfin.com/buy-a-home/comparative-market-analysis FOR INFO ON PRICING)'], inplace=True)
MN_cleaned.isna().sum()
```

```python
# determining which columns have only one value or only a couple values

MN_cleaned.nunique()
```

```python
# will check to see waht the single unique values are for specific columns

singles_subset = MN_cleaned[['SALE TYPE', 'STATE OR PROVINCE', 'STATUS', 'SOURCE', 'FAVORITE', 'INTERESTED']]
print(singles_subset.apply(pd.unique))
```

```python
# All of these columns can be removed since every value has this information

MN_cleaned.drop(columns=['SALE TYPE', 'STATE OR PROVINCE', 'STATUS', 'SOURCE', 'FAVORITE', 'INTERESTED'], inplace=True)
MN_cleaned.head()
```

```python
MN_cleaned.nunique()
```

```python
MN_cleaned.isna().sum()
```

```python
MN_cleaned.describe()
```

```python
# There are no HOA values of 0 which would be expected to be a significant number of the values.
# When uploading, it is likely the 0 values were transformed into NaN

MN_cleaned['HOA/MONTH'].fillna(0, inplace=True)
MN_cleaned.isna().sum()
```

```python
# 5 columns are all missing 13 values. 
# will test to see if they overlap by just searching for 1 set of NaN values

missing13 = MN_cleaned.loc[MN_cleaned['BEDS'].isna(),:]
missing13
```

```python
# all NaN values for these 5 columns are matched. This may be due to them being vacant land and therefore, they would not have beds or baths or a building.
# This is likely the only vacant land values, and considering this is a housing cost predictor, these values will be removed. 
# Also, there isnt much data on vacant land so any kind of model may not be a good predictor

print(MN_cleaned['PROPERTY TYPE'].value_counts())
```

```python
# The only vacant land values are the same as the missing 13 samples. Vacant land will be removed

MN_cleaned = MN_cleaned[MN_cleaned['PROPERTY TYPE'] != 'Vacant Land']
MN_cleaned.isna().sum()
```

```python
# Exploring Lot size = NaN

lot_size_0 = MN_cleaned[MN_cleaned['LOT SIZE'].isna()]
lot_size_0
```

```python
# The lot size is likely related to the property type being a condo unit or something where the land is not part of the deal.

lot_size_0['PROPERTY TYPE'].value_counts()
```

```python
lot_size_0['ADDRESS'].value_counts()
```

```python
MN_cleaned['PROPERTY TYPE'].value_counts()
```

```python
# Calculating the percent of each value that is NaN vs not

lot_size_0['PROPERTY TYPE'].value_counts() / MN_cleaned['PROPERTY TYPE'].value_counts() *100
```

```python
# These NaN values are likely 0 or close to 0. Condos and Townhouses make up the majority of these and may not actually include much land.
# This is where a conversation with a real estate agent would come in handy - will move forward assuming 0

MN_cleaned['LOT SIZE'].fillna(0, inplace=True)
MN_cleaned.isna().sum()
```

```python
# no NaNs exist, now to check length
MN_cleaned.info()
```

```python
# All columns are the same length, now to change the type

column_types = { 'PROPERTY TYPE': 'string',
                'ADDRESS': 'string',
                'CITY': 'string',
                'ZIP OR POSTAL CODE': 'string',
                'PRICE': 'float',
                'BEDS': 'float',
                'BATHS': 'float',
                'LOCATION': 'string',
                'SQUARE FEET': 'float',
                'LOT SIZE': 'float',
                'YEAR BUILT': 'int',
                'DAYS ON MARKET': 'int',
                '$/SQUARE FEET': 'float',
                'HOA/MONTH': 'float',
                'MLS#': 'string',
                'LATITUDE': 'float',
                'LONGITUDE': 'float'}

MN_cleaned = MN_cleaned.astype(column_types)
MN_cleaned.info()

```

```python
MN_cleaned.head()
```

```python
MN_cleaned.tail()
```

```python
# data is looking genrally good. will fix index.

MN_cleaned.reset_index(drop=True).head()
```

```python
# Checking for remaining duplicates
duplicated_row = MN_cleaned['ADDRESS'].duplicated().sum()
print(duplicated_row)
```

```python
# 1 duplicate found. Will remove

MN_cleaned.drop_duplicates(inplace=True)
MN_cleaned.duplicated().sum()
```

```python
# now to look for outliers. This will be done with a box plot 

fig, axes = plt.subplots(3, 3, figsize=(15,15))

axes = axes.flatten()

sns.boxplot(data=MN_cleaned, y='PRICE', ax=axes[0], color='blue')
axes[0].set_title('Price Distribution')

sns.boxplot(data=MN_cleaned, y='BEDS',ax=axes[1], color='blue')
axes[1].set_title('Beds')

sns.boxplot(data=MN_cleaned, y='BATHS',ax=axes[2], color='blue')
axes[2].set_title('Baths')

sns.boxplot(data=MN_cleaned, y='SQUARE FEET',ax=axes[3], color='blue')
axes[3].set_title('Square feet')

sns.boxplot(data=MN_cleaned, y='LOT SIZE',ax=axes[4], color='blue')
axes[4].set_title('Lot Size')

sns.boxplot(data=MN_cleaned, y='YEAR BUILT',ax=axes[5], color='blue')
axes[5].set_title('Year Built')

sns.boxplot(data=MN_cleaned, y='DAYS ON MARKET',ax=axes[6], color='blue')
axes[6].set_title('Days on Market')

sns.boxplot(data=MN_cleaned, y='$/SQUARE FEET',ax=axes[7], color='blue')
axes[7].set_title('Price per Square foot')

sns.boxplot(data=MN_cleaned, y='HOA/MONTH',ax=axes[8], color='blue')
axes[8].set_title('HOA')
                  
plt.show()
```

```python
# The housing price range does seem reasonable and even the more expensive houses have other points nearby suggesting they are real
fig, axes = plt.subplots(3, 3, figsize=(15,15))

axes = axes.flatten()

sns.histplot(data=MN_cleaned['PRICE'], ax=axes[0], color='blue', bins=30)
axes[0].set_title('Price Distribution')

sns.histplot(data=MN_cleaned['BEDS'], ax=axes[1], color='blue', bins=30)
axes[1].set_title('Beds')

sns.histplot(data=MN_cleaned['BATHS'], ax=axes[2], color='blue', bins=30)
axes[2].set_title('Baths')

sns.histplot(data=MN_cleaned['SQUARE FEET'], ax=axes[3], color='blue', bins=30)
axes[3].set_title('Square feet')

sns.histplot(data=MN_cleaned['LOT SIZE'], ax=axes[4], color='blue', bins=30)
axes[4].set_title('Lot Size')

sns.histplot(data=MN_cleaned['YEAR BUILT'], ax=axes[5], color='blue', bins=30)
axes[5].set_title('Year Built')

sns.histplot(data=MN_cleaned['DAYS ON MARKET'], ax=axes[6], color='blue', bins=30)
axes[6].set_title('Days on Market')

sns.histplot(data=MN_cleaned['$/SQUARE FEET'], ax=axes[7], color='blue', bins=30)
axes[7].set_title('Price per Square foot')

sns.histplot(data=MN_cleaned['HOA/MONTH'], ax=axes[8], color='blue', bins=30)
axes[8].set_title('HOA')
                  
plt.show()
```
