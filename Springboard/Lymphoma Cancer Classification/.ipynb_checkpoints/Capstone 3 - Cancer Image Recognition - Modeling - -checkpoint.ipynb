```python
# now to load the data

X_train = np.load('X_train_purified.npy')
y_train = np.load('y_train_purified.npy')
X_test = np.load('X_test_purified.npy')
y_test = np.load('y_test_purified.npy')
```

```python
class PCamExhaustiveDataset(Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels
        self.to_tensor = transforms.ToTensor()

    def __len__(self):
        # Every 1 physical image becomes 8 virtual images
        return len(self.images) * 8

    def __getitem__(self, idx):
        original_idx = idx // 8
        view_type = idx % 8
        
        # 1. Pull the 96x96 image from the SSD (mmap)
        img = self.images[original_idx]
        label = self.labels[original_idx]
        
        # 2. Basic Tensor Conversion
        img_tensor = self.to_tensor(img)
        
        # 3. The "Full Circle + Flip" Logic
        if view_type < 4:
            # Views 0, 1, 2, 3: Rotate 0, 90, 180, 270 degrees
            img_tensor = torch.rot90(img_tensor, k=view_type, dims=[1, 2])
        else:
            # Views 4, 5, 6, 7: Flip first, then Rotate 0, 90, 180, 270
            img_tensor = torch.flip(img_tensor, dims=[2]) 
            img_tensor = torch.rot90(img_tensor, k=(view_type - 4), dims=[1, 2])
            
        return img_tensor, label
```

```python
import matplotlib.pyplot as plt

# 1. Grab one base image from your RAM-loaded X_train
sample_idx = 0 
base_img = X_train[sample_idx]
label = y_train[sample_idx]

# 2. Plot the 8-way Exhaustive Symmetry
plt.figure(figsize=(16, 8))

for i in range(8):
    plt.subplot(2, 4, i + 1)
    
    # Logic: 0-3 are rotations, 4-7 are flipped then rotations
    if i < 4:
        # np.rot90 works on (H, W, C), k is number of 90-degree rotations
        view = np.rot90(base_img, k=i)
        title = f"Rot {i*90}°"
    else:
        # Flip left-to-right, then rotate
        view = np.rot90(np.fliplr(base_img), k=i-4)
        title = f"Flip + Rot {(i-4)*90}°"
        
    plt.imshow(view)
    plt.title(title)
    plt.axis('off')

plt.suptitle(f"Exhaustive 8-Way Symmetry (Target: {'Cancer' if label==1 else 'Healthy'})", fontsize=16)
plt.tight_layout()
plt.show()
```

# Model 1: NSFW Image Clasifier

```python
import torch
import torch.nn as nn
import numpy as np
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import torchvision.transforms.functional as TF
from transformers import ViTConfig, ViTForImageClassification

# 1. FORCE CPU (since the Blackwell drivers are still broken)
device = torch.device("cpu")

# 2. Reload data into RAM (This is fast on an Ultra 9)
X_train = np.load('X_train_purified.npy')
y_train = np.load('y_train_purified.npy')

# 3. Setup the High-Recall Weights
loss_weights = torch.tensor([1.0, 2.0]).to(device)
criterion = nn.CrossEntropyLoss(weight=loss_weights)
vit_resize = transforms.Resize((224, 224), interpolation=transforms.InterpolationMode.BICUBIC)

print(f"✅ Context Restored. Ready for training on {device}.")
```

```python
class PCamExhaustiveDataset(Dataset):
    def __init__(self, X_array, y_array):
        self.X = X_array
        self.y = y_array
        self.real_len = len(self.X)

    def __len__(self):
        return self.real_len * 8

    def __getitem__(self, idx):
        original_idx = idx // 8
        transform_id = idx % 8
        
        image = self.X[original_idx]
        label = self.y[original_idx]
        
        image = torch.from_numpy(image).permute(2, 0, 1).float() / 255.0
        label = torch.tensor(label).long()

        if transform_id < 4:
            return TF.rotate(image, transform_id * 90), label
        else:
            flipped = TF.hflip(image)
            return TF.rotate(flipped, (transform_id - 4) * 90), label

# Instantiate the pipeline
train_ds = PCamExhaustiveDataset(X_train, y_train)
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True, num_workers=0)

# Re-init Model and Optimizer
config = ViTConfig(num_labels=2, image_size=224, patch_size=16)
model_vit = ViTForImageClassification(config).to(device)
optimizer = torch.optim.AdamW(model_vit.parameters(), lr=5e-5)
```

```python
import time

MAX_BATCHES = 15000 
model_vit.train()
start_time = time.time()

try:
    for i, (inputs, labels) in enumerate(train_loader):
        if i >= MAX_BATCHES: break
            
        inputs = vit_resize(inputs).to(device)
        labels = labels.to(device)
        
        optimizer.zero_grad()
        outputs = model_vit(inputs).logits
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        
        if i % 10 == 0:
            elapsed = (time.time() - start_time) / 60
            print(f"Batch {i}/{MAX_BATCHES} | Loss: {loss.item():.4f} | {elapsed:.1f}m", end='\r')

finally:
    torch.save(model_vit.state_dict(), 'falconsai_vit_milestone_15k.pth')
    print("\n✅ Session Resumed and Final Weights Protected.")
```

```python
import torch
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Prep Test Data
# Assuming you have X_test and y_test in RAM from your earlier cleanup
class PCamTestDataset(Dataset):
    def __init__(self, X, y):
        self.X, self.y = X, y
    def __len__(self): return len(self.X)
    def __getitem__(self, idx):
        img = torch.from_numpy(self.X[idx]).permute(2, 0, 1).float() / 255.0
        return vit_resize(img), torch.tensor(self.y[idx]).long()

test_ds = PCamTestDataset(X_test, y_test)
test_loader = DataLoader(test_ds, batch_size=32, shuffle=False)

# 2. Run Inference
model_vit.eval()
all_preds, all_labels = [], []

print("Evaluating Recall on Test Set...")
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.to(device)
        outputs = model_vit(inputs).logits
        preds = torch.argmax(outputs, dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())

# 3. The Money Shot: Metrics
print("\n--- PERFORMANCE PROFILE ---")
print(classification_report(all_labels, all_preds, target_names=['Healthy', 'CANCER']))

# 4. Plot Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', xticklabels=['Healthy', 'CANCER'], yticklabels=['Healthy', 'CANCER'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Clinical Recall Validation')
plt.show()
```

```python
import torch.nn as nn

# Weights moved to CPU
loss_weights = torch.tensor([1.0, 2.0]).to(device)
criterion = nn.CrossEntropyLoss(weight=loss_weights)

# ViT standard resizer
vit_resize = transforms.Resize((224, 224), interpolation=transforms.InterpolationMode.BICUBIC)
```

```python
# 1. Instantiate the Exhaustive Dataset (8x virtual images)
# Using the X_train and y_train variables currently in your RAM
train_ds = PCamExhaustiveDataset(X_train, y_train)

# 2. Build the Loader
# batch_size 32 is a safe "sweet spot" for CPU processing
train_loader = DataLoader(
    train_ds, 
    batch_size=32, 
    shuffle=True, 
    num_workers=0
)

print(f"✅ train_loader defined.")
print(f"Total virtual training samples: {len(train_ds):,}")
print(f"Total batches to process: {len(train_loader):,}")
```

```python
from transformers import ViTConfig, ViTForImageClassification

# Initialize Falconsai ViT from scratch for the experiment
config = ViTConfig(num_labels=2, image_size=224, patch_size=16)
model_vit = ViTForImageClassification(config).to(device)

print("✅ Falconsai ViT initialized and ready on CPU.")
```

```python
import torch.optim as optim

# Standard learning rate for ViT fine-tuning
optimizer = optim.AdamW(model_vit.parameters(), lr=5e-5)
print("✅ Optimizer linked to Model.")
```

```python
import time

# 1. Config for the Early Exit
MAX_BATCHES = 15000  # Approximately 480k virtual images
model_vit.train()

print(f"🚀 Training Milestone Set: {MAX_BATCHES} batches")
print(f"Target: High Recall (Weight 2.0 on Cancer) | Device: {device}")

start_time = time.time()
running_loss = 0.0

try:
    for i, (inputs, labels) in enumerate(train_loader):
        # Stop check
        if i >= MAX_BATCHES:
            print(f"\n🎯 Milestone reached at {i} batches. Stopping for evaluation.")
            break
            
        # Step A: Prep
        inputs = vit_resize(inputs).to(device)
        labels = labels.to(device)
        
        # Step B: Forward & Loss
        optimizer.zero_grad()
        outputs = model_vit(inputs).logits
        loss = criterion(outputs, labels)
        
        # Step C: Backward (The Ultra 9 heavy lifting)
        loss.backward()
        optimizer.step()
        
        # Step D: Stats
        running_loss += loss.item()
        if i % 10 == 0:
            elapsed = (time.time() - start_time) / 60
            remaining = (elapsed / (i+1)) * (MAX_BATCHES - i)
            print(f"Batch {i}/{MAX_BATCHES} | Loss: {loss.item():.4f} | Est. Remaining: {remaining:.1f}m", end='\r')
            
        # Emergency Checkpoint every 5k batches
        if i % 5000 == 0 and i > 0:
            torch.save(model_vit.state_dict(), f'vit_recall_checkpoint_{i}.pth')

except Exception as e:
    print(f"\n❌ Error encountered: {e}")

finally:
    # This is your Capstone Gold
    save_path = 'falconsai_vit_milestone_15k.pth'
    torch.save(model_vit.state_dict(), save_path)
    print(f"\n✅ Weights saved to '{save_path}'")
    print(f"Total time elapsed: {(time.time() - start_time)/3600:.2f} hours")
```
